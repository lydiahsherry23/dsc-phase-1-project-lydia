```python
#Display all columns
pd.set_option('display.max.columns', None)

#percentage of null values in the columns
null_percentage = (df.isnull().sum()/len(df)*100).round(2)
null_percentage

#convert to dataframe
null_percentage_df = null_percentage.to_frame(name='Null Percentage')
null_percentage_df
```

```python
#columns list that has a lower percentage of null values
#columns = ["Event.Id","Accident.Number","Event.Date","Location","Country","Injury.Severity","Aircraft.damage","Registration.Number","Make","Model","Amateur.Built","Number.of.Engines","Engine.Type","Purpose.of.flight","Total.Uninjured","Weather.Condition","Report.Status"]
columns = [key for key ,value in null_percentage.items() if value < 10]
#drop rows that have null values in the columns
df.dropna(subset = columns,inplace = True)
```

```python
df.isnull().sum()
```

```python
#fill in aircraft category column to predict the category,if not unknown
def fill_nan_aircraft_category(df):
    df['Aircraft.Category'] = df.groupby('Make')['Aircraft.Category'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
    return df
fill_nan_aircraft_category(df)
```

### DATA EXPLORATION

### distribution of accidents over a year

```python
#Visualize number of accidents against year 
#Convertion and extraction of 'Event.Date' to datetime format 
df['Event.Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Event.Date'].dt.year

# accident counts per yr
accidents_per_year = df['Year'].value_counts().sort_index()

#data plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=accidents_per_year.index, y=accidents_per_year.values)
plt.title('Number of Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

we observe the following from the above markdown

1. there were no aeroplanes before 1980
2. year 1980 recorded the highest accidents
3. number of accidents are reducing as the years go by due to improvement of skilled manpower and technology

### distribution of accidents wit aircraft category

```python
plt.figure(figsize=(12, 6))
#create a seaborn plot
sns.countplot(data=df, x='Aircraft.Category', order=df['Aircraft.Category'].value_counts().index)
#graph title
plt.title('Distribution of Accidents by Aircraft Category')
#x label
plt.xlabel('Aircraft Category')
#y label
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.yscale('log')
plt.show()
```

```python
df.groupby('Aircraft.Category')['Make'].count()
```

we can note that airplane has the highest number of accidents but generally the number of accidents by category is ina descending order

### distribution of injuries by broad phase of flight

```python
phase_of_flight_data = df.groupby('Broad.phase.of.flight').agg(
    total_incidents=('Event.Id', 'count'),
    total_fatal_injuries=('Total.Fatal.Injuries', 'sum'),
    total_serious_injuries=('Total.Serious.Injuries', 'sum'),
    total_minor_injuries=('Total.Minor.Injuries', 'sum')
).reset_index()
plt.figure(figsize=(14, 8))
plt.subplot(2, 1, 1)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_incidents'], color='blue')
plt.title('Total Incidents by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Incidents')
plt.xticks(rotation=45)
plt.grid(True)
plt.subplot(2, 1, 2)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_fatal_injuries'], color='red', alpha=0.5, label='Fatal Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_serious_injuries'], color='orange', alpha=0.5, label='Serious Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_minor_injuries'], color='green', alpha=0.5, label='Minor Injuries')
plt.title('Total Injuries by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Injuries')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

two subplots are created

1. first subplot shows the total number of accidents by phase of flight
2. second subplot shows the total number of fatal,serious and minor injuries by phase of flight differentiated by color
